```python
import pandas as pd
import numpy as np
import sqlite3
```

```python
con = sqlite3.connect("./movies.db")
```

```python
id = [i for i in range(12018)]
id = list(map(lambda x: '{0:0>5}'.format(x), id))
```

```python
audience_list = []
play_list = []

for i in id:
    tmp = pd.read_sql("SELECT * FROM "+ "'" + i + "'", con, index_col=None)

    audience = pd.to_numeric(tmp['audience'].str.replace(',',''), errors='coerce').sum()
    play = pd.to_numeric(tmp['play'].str.replace(',',''), errors='coerce').sum()
    
    audience_list.append(audience)
    play_list.append(play)
```

```python
features = pd.read_sql("SELECT * FROM features", con, index_col=None)
features['accum_audience'] = audience_list
features['accum_play'] = play_list
```

```python
features
```

```python
filtered = features[(features['accum_audience'] >= 10000)]
filtered
```

```python
df = pd.read_csv('./data/score.csv')
df
```

```python
score = pd.DataFrame({'title': df['영화명'], 'score': df['평점']})
# score.index = df['영화명']
score
```

```python
filtered['score']= np.nan

col_score = []

for i in filtered.index:
    lookup = filtered['title'][i] == score['title']
    try:
        col_score.append(score[lookup]['score'].values.item())
    except:
        col_score.append(np.nan)

#     print(score[lookup]['score'].values.item())
#     print(score[lookup]['score'].values)
#     if lookup.any():
#         filtered['score'][i] = score[lookup]['score']
```

```python
col_score
```

```python
filtered['score'] = col_score
filtered
```

```python
filtered.to_sql('features', con, index=False, if_exists='replace')
```
