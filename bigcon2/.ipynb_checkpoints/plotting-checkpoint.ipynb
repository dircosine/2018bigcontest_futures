```python
con = sqlite3.connect("./movies.db")
```

```python
features = pd.read_sql("SELECT * FROM features", con, index_col=None)
over_ = features[(features['open'] >= '2015-01-01')]
over_
```

```python
# end_date = datetime.datetime(2018, 8, 15)

# index = pd.date_range(end_date-datetime.timedelta(224), periods=224, freq='D')
# columns = []

# df_merged = pd.DataFrame(index=index, columns=columns)
# df_merged = df_merged.fillna(0)


audience_merged = pd.DataFrame({})
play_merged = pd.DataFrame({})

for i in over_.index:
#     if i == 51:
#         continue
        
    print(i)
        
    tmp = pd.read_sql("SELECT * FROM " + '"' + over_.loc[i]['id'] + '"', con, index_col='date')
    
    title = over_.loc[i]['title']
    
    audience = pd.to_numeric(tmp['audience'].str.replace(',',''), errors='coerce')
    audience.index = pd.to_datetime(audience.index)
    audience = pd.DataFrame({title: audience})
    audience_merged = pd.concat([audience_merged, audience], axis=1)

    play = pd.to_numeric(tmp['play'].str.replace(',',''), errors='coerce')
    play.index = pd.to_datetime(play.index)
    play = pd.DataFrame({title: play})
    play_merged = pd.concat([play_merged, play], axis=1)
    

audience_merged
```

```python
audience_merged['total'] = audience_merged.sum(axis=1)
audience_merged
```

```python
df_gantt = []
for i in over_.index:
    tmp = pd.read_sql("SELECT * FROM " + '"' + over_.loc[i]['id'] + '"', con, index_col=None)
    
#     tmp = tmp.sort_values(by=['date'])
    
    task = over_.loc[i]['title']
    open = over_.loc[i]['open']
    start = list(tmp['date'])[0]
    finish = list(tmp['date'])[-1]
    
    df_gantt.append(dict(Task=task, Start=start, Finish=finish))
    
df_gantt
```
