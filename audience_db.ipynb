```python
import pandas as pd
import datetime as dt
import numpy as np
import sqlite3
```

```python
con = sqlite3.connect("./movies.db")
title = pd.read_sql("SELECT * FROM features", con, index_col=None)
title
```

```python
wd = './data/2018/'
```

```python
 for file in range(36):
    df = pd.read_csv(wd + str(file) + '.csv')
    print(file)

    date = [df.loc[3][i] for i in range(11,48,6)]

    for _df in df[5:-1].iterrows():    
        lookup = title['title'] == _df[1][1]
#         idx = title[lookup].index.format()[0]
        id = title[lookup]['id'].values[0]
        
        audience = [_df[1][i] for i in range(13,50,6)]
        play = [_df[1][i] for i in range(16,53,6)]

        tmp = pd.DataFrame({'date': date, 'audience': audience, 'play': play})
        tmp = tmp[(tmp.T != '0').all()]

        tmp.to_sql(id, con, index=False, if_exists='append')
```
