```python
import pandas as pd
import datetime as dt
import numpy as np
import sqlite3
```

```python
empty = {'Unnamed: 1': [],'Unnamed: 2': [],'Unnamed: 3': [],'Unnamed: 5': [], 'Unnamed: 6': [],
                    'Unnamed: 7': [],'Unnamed: 8': [],'Unnamed: 9': [],'Unnamed: 10': []}


df = pd.DataFrame(empty)
```

```python
wd = './data/2018/'
```

```python
for i in range(36):
    print(i)
    file_name = str(i) + '.csv'

    feature_cols = ['Unnamed: 1','Unnamed: 2','Unnamed: 3','Unnamed: 5', 'Unnamed: 6',
                    'Unnamed: 7','Unnamed: 8','Unnamed: 9','Unnamed: 10']

    _df = pd.read_csv(wd + file_name)
    df = df.append(_df[feature_cols][5:-1], ignore_index=True)
    df = df.drop_duplicates(['Unnamed: 1'], keep='first')

df
```

```python
col_names = ['title', 'open', 'nation', 'producer', 'distributor', 
         'grade', 'genre', 'director', 'actors']

df.columns = col_names

df
```

```python
id = range(0, len(df))
# del df['id']
df.insert(loc=0, column='id', value=id)
df['id'] = df['id'].apply(lambda x: '{0:0>5}'.format(x))
df
# index
```

```python
con = sqlite3.connect('./movies.db')

df.to_sql('features', con, index=False, if_exists='replace')
```
